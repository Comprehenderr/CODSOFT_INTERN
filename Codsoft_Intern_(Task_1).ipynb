```python
import random
```

```python
# Define responses based on predefined rules
responses = {
    "hi": "Hello! How can I assist you today?",
    "hello": "Hi there! What can I do for you?",
    "how are you?": "I'm just a bot, but I'm here to help. How can I assist you?",
    "bye": "Goodbye! Have a great day.",
    "thanks": "You're welcome!",
    "default": "I'm sorry, I didn't understand that. Can you please rephrase or ask something else?",
}
```

```python
# Function to provide responses
def chatbot_response(user_input):
    input_text = user_input.lower()  # Convert user input to lowercase for case insensitivity

    # Check if the input matches any predefined rules
    if input_text in responses:
        return responses[input_text]
    else:
        return responses["default"]
```

```python
# Main function for chatting
def chat():
    print("Welcome! Ask me anything or say 'bye' to exit.")
    while True:
        user_message = input("You: ")
        if user_message.lower() == 'bye':
            print("Bot: Goodbye! Have a great day.")
            break
        else:
            bot_response = chatbot_response(user_message)
            print("Bot:", bot_response)
```

```python
# Start the conversation
chat()
```
